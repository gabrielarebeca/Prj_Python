Importa as bibliotecas "pandas" "matplotlib.pyplot" e "seaborn"
import as é só para informar que o import "biblioteca" vai ter um acrônimo como "LB"

```python
import pandas as pd #encurta a biblioteca de software (pandas) para pd
import matplotlib.pyplot as plt #encurta a biblioteca de visualização de dados (matplotlib) para plt
import seaborn as sns #encurta biblioteca de visualização de dados (seaborn) para sns
```

transforma a tabela .CSV em variavel 
#latin-1 é para os caracteres especiais (ç, ã)
#sep é a forma como o essa lista expecifica identifica os campos da tabela

```python
df = pd.read_csv('lista_de_espera_sisu_2022_2.csv', sep='|', encoding='latin-1') #transforma a planilha na variavel df
```

Verifica se a leitura da tabela está funcionando

Faz as leituras iniciais

```python
print(df.head(10)) #vai mostrar apenas os 10 primeiros dados da tabela
print(df.columns) #vai mostrar as colunas dessas dados
print(df.columns [:10]) #vai mostrar as 10 primeiros colunas dessas dados
```

exibir campos e estado exibe campus selecionados

```python
campus_por_estado = df['UF_CAMPUS'].value_counts() #value_counts pela os valos que tem por campo
print(campus_por_estado)

cursos = df['NOME_CURSO'].value_counts()
print(cursos)
```

cria os graficos tipo "PIE" ou "pizza" e o bar como "BAR" ou "lista"

```python
campus_por_estado.plot(kind='bar')
plt.xlabel('UF_CAMPUS')
plt.title('Distribuição de Campus por Estado')
plt.show(10) #### **VER SE VAI CONTAR ATÉ 10

top_10_cursos = cursos.sort_values(ascending=False).head(10) #esse codigo mostra os primeiros 10 da tabela
top_10_cursos.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribuição de Cursos')
plt.show(10) #### **VER SE VAI CONTAR ATÉ 10
```

Importa a biblioteca fpdf2

```python
from fpdf import FPDF
```

Cria o PDF dessa exibição de gráficos

```python
#inicia o FPDF e quebra de linha automatica
pdf = FPDF()
pdf.set_auto_page_break(auto=True, margin=15)
pdf.add_page()
```

```python
#definições de texto
pdf.set_font("Arial", size=12)
pdf.cell(0, 10, txt="Relatório de Análise do SISU 2022", ln=True, align='C')
```

```python
#tipo de grafico
campus_por_estado_plot = 'campus_por_estado.png'
campus_por_estado.plot (kind='bar')
plt.xlabel('UF_CAMPUS')
plt.title('Distribuição de Campus por Estado')
plt.savefig (campus_por_estado_plot)
plt.close()

cursos_plot = 'cursos.png'
cursos.plot (kind='pie', autopct='%1.1f%%')
plt.title('Distribuição de Cursos')
plt.savefig (cursos_plot)
plt.close()
```

```python
#gerar o PDF
pdf = FPDF ()
pdf.set_auto_page_break(auto=True, margin=15)
pdf.add_page()
```

```python
#editar fonte
pdf. set_font('Arial', 'B', size=12)
```

```python
#cria os gráficos dentro do pdf
pdf.add_page()
pdf.set_font('Arial', 'B', size=12)
pdf.cell(0, 10, 'Gráficos', ln=True, align='C')

pdf.image(campus_por_estado_plot, x=10, y=30, w=pdf.w - 20)
pdf.add_page()
pdf.image(cursos_plot, x=10, y=30, w=pdf.w - 20)
```

```python
#finaliza e cria o PDF
pdf.output ('GABI.pdf')
print ("PDF gerado com sucesso!")
```
