```python
df = pd.read_csv('lista_de_espera_sisu_2022_2.csv', sep='|', encoding='latin-1') #transforma a planilha na variavel df
```

#latin-1 é para os caracteres especiais (ç, ã)
#sep é a forma como o essa lista expecifica identifica os campos da tabela


```python
print(df.head())
print(df.columns)
```

faz as leituras

```python
campus_por_estado = df ['UF_CAMPUS'].value_counts() #value_counts pela os valos que tem por campo
print (campus_por_estado)

cursos = df ['NOME_CURSO'].value_counts()
print (cursos)
```

exibir campos e estado

```python
campus_por_estado.plot (kind='bar')
plt.xlabel ('UF_CAMPUS')
plt.title('Distribuição de Campus por Estado')
plt.show ()

top_10_cursos = cursos.sort_values (ascending=False).head(10) #esse codigo mostra os primeiros 10 da tabela
top_10_cursos.plot (kind='pie', autopct='%1.1f%%')
plt.title ('Distribuição de Cursos')
plt.show()
```

cria os graficos tipo "PIE" ou "pizza" e o bar como "BAR" ou "lista"
